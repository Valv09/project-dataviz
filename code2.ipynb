```python
conteo = df[['name', 'genre']].explode('genre')
```

```python
conteo['Count'] = 1
conteo.head(5)
```

```python
pivote_genero = pd.pivot_table(conteo,
                        index='genre',
                        aggfunc={'Count': 'count'},
).sort_values('Count', ascending=False)
pivote_genero
```

```python
df_genero = conteo.groupby('genre', as_index=False).agg({'Count': 'count'}).sort_values('Count', ascending=False)
df_genero
```

```python
plt.figure(figsize=(15, 8))
ax = sns.barplot(data=df_genero, x='genre', y='Count', color='mediumorchid')
ax.set_xticklabels(ax.get_xticklabels(), rotation=65);
```

```python
plt.figure(figsize=(12, 5))
sns.set_theme(style='darkgrid')
sns.color_palette("ch:s=.25,rot=-.25", as_cmap=True)
sns.countplot(data=df_genero, x='genre')
plt.xticks(rotation=60);
```

```python
mask = np.zeros_like(df.corr(), dtype='bool')
mask
```

```python
mask[np.triu_indices_from(mask)] = True
mask
```

```python
plt.subplots(figsize=(15, 10))
cmap = sns.diverging_palette(255, 6, as_cmap=True)
figure = sns.heatmap(df.corr(), mask=mask, cmap=cmap,
                     center=0, linewidths=0.5, 
                     annot=True, fmt='.2f'
                     )
```
